```python
import emoji

def extract_emojis(s):
  return [c for c in s if c in emoji.UNICODE_EMOJI['en']]

# Ejemplo:
sentence = "Feeling üòä after such a long day! But tomorrow's schedule looks ü§î and the weather is üåßÔ∏è. I'm traveling ‚úàÔ∏è to üóΩ next week üöó. I love traveling üåç‚úàÔ∏èüö¢ and exploring new countries üó∫Ô∏è"
emojis_in_sentence = extract_emojis(sentence)
print(emojis_in_sentence)  # Output: ['üòä', 'ü§î', 'üåß', '‚úà', 'üóΩ', 'üöó', 'üåç', '‚úà', 'üö¢', 'üó∫']
```

```python
def calculate_sentiment(text, emoji_lexicon):
    #Tokenize and Identify Emojis
    emojis_in_text = extract_emojis(text)  # Function to extract emojis from text
    sentiment_scores = [emoji_lexicon.get(emoji, 0) for emoji in emojis_in_text]

    #Aggregate Scores
    aggregated_score = sum(sentiment_scores)

    #Perform Textual Sentiment Analysis (if applicable)
    textual_sentiment_score = 0
    #Combine Emoji and Textual Sentiment
    combined_score = aggregated_score + textual_sentiment_score

    #Thresholding
    if combined_score > 0.5:
        sentiment = "Positive"
    elif combined_score < -0.25:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    return sentiment
```

```python

# Sample Lexicon
emoji_lexicon = {
    "üòä": 0.8,   # Positive
    "üò¢": -0.7,  # Negative
    "ü§î": 0.1,   # Neutral
    "üëç": 0.9,   # Positive
    "üëé": -0.9,  # Negative
    "üòÑ": 0.8,   # Positive
    "üòû": -0.6,  # Negative
    "üòç": 0.9,   # Positive
    "üò°": -0.8   # Negative
}
```

```python
import regex as re

def get_emoji_frequency(dataframe, column_name):
    emojis_dict = {}
    
    # Function to extract emojis from text
    def extract_emojis_from_text(text):
        return re.findall(r'\p{So}', text)
    
    # Apply the function to each row in the specified column
    emojis_lists = dataframe[column_name].apply(extract_emojis_from_text)
    
    # Iterate through each list of emojis and count their occurrences
    for emojis_list in emojis_lists:
        for emoji in emojis_list:
            if emoji in emojis_dict:
                emojis_dict[emoji] += 1
            else:
                emojis_dict[emoji] = 1
    
    # Create a DataFrame from the dictionary of emojis and their frequencies
    emojis_df = pd.DataFrame(list(emojis_dict.items()), columns=['Emoji', 'Frequency'])
    
    # Sort the DataFrame by frequency in descending order
    emojis_df = emojis_df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
    
    return emojis_df
```

```python
emojis_df = get_emoji_frequency(df, 'text')
print(emojis_df)
```
