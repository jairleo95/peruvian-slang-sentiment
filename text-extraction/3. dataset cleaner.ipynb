```python
import pandas as pd
```

```python
df = pd.read_csv(r'[labeled]dataset-01.csv',usecols=['id_slang', 'slang', 'text', 'label', 'char_count'], encoding='latin-1')
df
```

```python
df = df[3300:]
```

```python
df = df.drop_duplicates(subset=['text'], keep='first')
```

```python
df.sample(30)
```

```python
characters = ['EL CHACAL:', 'LA LOCA:', 'CARADURA:', 'LA PERIODISTA', 'LUCÍA:', 'CLAUDIA:', 'CARLOS:', 'TYSON:', 'KILLER:']

def remove_characters(text):
    for x in characters:
        text = text.replace(x, '')
    return text
```

```python
df["text"] = df["text"].apply(lambda x: remove_characters(x))
df
```
