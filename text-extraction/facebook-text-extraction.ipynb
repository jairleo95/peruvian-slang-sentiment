```python
#exclude words
words = ['jaja','jajaja', 'jajajajajajja', 'jajajajajajaja','jajajajjajajajaja','Jajajajajajajajaja','jajajaja','jajajajjajajajaja', 'jajajajja','jijiji', 'saludo', 'saluda', 'xd', 'hola', 'zzz']
def remove_words(text):
    for x in words:
        text = text.lower().replace(x, '')
    if len(text)>=MIN_CHAR_LENGHT:
        return True
    return False
```

```python
SENTENCE_LENGTH = 10000000 #limit of characters to extract
def extract_sentence(corpus, text):
    sentences = []
    for line in list(corpus):
        if (" "+text in line.lower()) & (len(line) <= SENTENCE_LENGTH):
#             print(line)
            sentences.append(line.strip().replace("\n"," "))
#             print('=============================================================================')
        
#     print(sentences)
    return sentences
```

```python
def find_slang(text):
    slangs = ""
    for i in df_slang.index:
        slang = df_slang['Jerga'][i].lower()
        if (len(slang)>slang_length):
            if " "+slang in text.lower():
                slangs += slang + ","
    return slangs
```
