```python
import sys
import pandas as pd 
import os 
import glob 
```

```python
csv_files = glob.glob(os.path.join(r"facebook/extraction-*/", "*.csv"))
```

```python
csv_files
```

```python
df_list = (pd.read_csv(file, sep = ';') for file in csv_files)
```

```python
big_df = pd.concat(df_list, ignore_index=True)
```

```python
df = big_df.dropna(subset=['message'])
```

```python
df = df[df['message'].str.len()>15]
```

```python
df.info()
```

```python
df = df.drop_duplicates(subset=['message'], keep='first')
```

```python
df
```

```python
#exclude words
words = ['jaja','jajaja', 'jajajajajajaja','jajajajjajajajaja','Jajajajajajajajaja','jajajaja','jajajajjajajajaja', 'jajajajja','jijiji', 'saludo', 'saluda', 'xd', 'hola', 'zzz']
min_length = 20

def remove_words(text):
    for x in words:
        text = text.lower().replace(x, '')
    if len(text)>min_length:
        return True
    return False
```

```python
print(remove_words('Bel茅n S谩nchez Castillo jajajajaja'))
```

```python
df
```

```python
#remove rows that are too small when excluding the list.
df['valid'] = df['message'].apply(lambda row: remove_words(row))
df
```
