## Loading books

```python
BASE_PATH = 'books/'

books = ["arguedas-Los ríos profundos.txt",
         "Conversación en La Catedral - Mario Vargas Llosa.txt",
        "el-mundo-es-ancho-y-ajeno-de-ciro-alegrc3ada.txt",
        "La-ciudad-y-los-perros-Vargas-LLosa.txt",
        "No se lo digas a nadie - Jaime Bayly.txt",
        "miyashiro-aldo-un-misterio-una-pasion-teatro-peruano.txt",
        "La-Noche-Es-Virgen-Jaime-Bayly.txt"]
```

```python
text_split = []
for book in books:
    print("book:", book)
    raw_text = open(BASE_PATH+book, 'r', encoding='utf-8').read()
    text_split +=raw_text.split('.')
```

## Slang Extraction

```python
#todo: filter the exact expression ' cachacho ', not in plural?

def extract_sentence(corpus, text):
    sentences = []
    for line in list(corpus):
        if (" "+text in line.lower()) & (len(line) <= SENTENCE_LENGTH):
#             print(line)
            sentences.append(line.strip().replace("\n"," "))
#             print('=============================================================================')
        
#     print(sentences)
    return sentences
```

```python
extract_sentence(text_split, "cachaco")
```

```python
data = []
slang_length = 3 #minimun length of characters of an slang
```

```python
#filter only slang words
for i in df.index:
    #if (df['tipo'][i] == 'Jerga') & (len(df['Jerga'][i]) >slang_length):
    if (len(df['Jerga'][i]) >slang_length):
        id_slang = int(df['nro'][i])
        slang = df['Jerga'][i].lower()
        sentences = extract_sentence(text_split, slang)
        
        for line in sentences:
            data.append({'id_slang': id_slang ,'slang': slang, 'sentence': line})
```

```python
df_raw = pd.DataFrame(data, columns=['id_slang','slang', 'sentence'])
df_raw
```
