```python
# load ascii text and covert to lowercase
filename = "el-mundo-es-ancho-y-ajeno-de-ciro-alegrc3ada.txt"
raw_text = open(filename, 'r', encoding='utf-8').read()
#raw_text = raw_text.lower()
```

```python
word = 'cholo'
```

```python
word in raw_text
```

```python
text_split = raw_text.split('.')
```

```python
text_split
```

```python
for line in list(text_split):
    if "cholo" in line:
        print(line.strip().replace("\n"," "))
        print('=============================================================================')
```

```python
for line in list(text_split):
    if "hueco" in line:
        print(line)
```

```python
for line in list(text_split):
    if "calato" in line:
        print(line)
```

```python
for line in list(text_split):
    if "chocar" in line:
        print(line)
        print(line.index("chocar"))
```
